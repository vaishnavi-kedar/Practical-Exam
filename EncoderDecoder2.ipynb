```python
# ===============================================
# 📌 Headline Generation using LSTM Variants (Final Optimized Version)
# ===============================================

# STEP 1: Import Libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import re
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense, Layer, Concatenate, RepeatVector
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.meteor_score import meteor_score
from rouge_score import rouge_scorer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

# STEP 2: Load and Prepare Dataset

def load_dataset():
    try:
        df1 = pd.read_csv('/content/news_summary.csv', encoding='ISO-8859-1')[['text', 'headlines']]
        df2 = pd.read_csv('/content/news_summary_more.csv', encoding='ISO-8859-1')[['text', 'headlines']]
        df = pd.concat([df1, df2]).dropna().reset_index(drop=True)
        df = df.sample(min(200, len(df)), random_state=42)
        if df.empty:
            raise ValueError("The dataset is empty after loading.")
        print(f"Dataset loaded successfully with {len(df)} records.")
        return df
    except Exception as e:
        print(f"Error loading dataset: {e}")
        topics = ['technology', 'sports', 'politics', 'health', 'business']
        samples = [{
            'text': f"Recent developments in {topic} show significant progress in area {i}",
            'headlines': f"{topic.capitalize()} advances in area {i}"
        } for topic in topics for i in range(20)]
        df = pd.DataFrame(samples)
        print("Using enhanced dummy dataset with variety")
        return df

df = load_dataset()

# STEP 3: Clean and Prepare Text Data

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

df['clean_text'] = df['text'].apply(clean_text)
df['clean_headlines'] = df['headlines'].apply(clean_text)
df['clean_headlines'] = 'start ' + df['clean_headlines'] + ' end'

# STEP 4: Tokenization and Padding

VOCAB_SIZE = 5000
MAX_INPUT_LEN = 50
MAX_TARGET_LEN = 15

input_tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<OOV>")
input_tokenizer.fit_on_texts(df['clean_text'])

target_tokenizer = Tokenizer(num_words=VOCAB_SIZE, oov_token="<OOV>")
target_tokenizer.fit_on_texts(df['clean_headlines'])

input_seq = pad_sequences(input_tokenizer.texts_to_sequences(df['clean_text']), maxlen=MAX_INPUT_LEN, padding='post')
target_seq = pad_sequences(target_tokenizer.texts_to_sequences(df['clean_headlines']), maxlen=MAX_TARGET_LEN, padding='post')

decoder_input_seq = target_seq[:, :-1]
decoder_target_seq = np.expand_dims(target_seq[:, 1:], -1)

X_train, X_val, Xd_train, Xd_val, Y_train, Y_val = train_test_split(
    input_seq, decoder_input_seq, decoder_target_seq, test_size=0.2, random_state=42
)

# STEP 5: Model Architectures

EMBEDDING_DIM = 256
LSTM_UNITS = 256

def build_lstm_model():
    encoder_inputs = Input(shape=(MAX_INPUT_LEN,))
    enc_emb = Embedding(VOCAB_SIZE, EMBEDDING_DIM)(encoder_inputs)
    enc_out, state_h, state_c = LSTM(LSTM_UNITS, return_state=True)(enc_emb)

    decoder_inputs = Input(shape=(MAX_TARGET_LEN-1,))
    dec_emb = Embedding(VOCAB_SIZE, EMBEDDING_DIM)(decoder_inputs)
    dec_lstm = LSTM(LSTM_UNITS, return_sequences=True)(dec_emb, initial_state=[state_h, state_c])
    output = Dense(VOCAB_SIZE, activation='softmax')(dec_lstm)

    model = Model([encoder_inputs, decoder_inputs], output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

class BahdanauAttention(Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = Dense(units)
        self.W2 = Dense(units)
        self.V = Dense(1)

    def call(self, query, values):
        query = tf.expand_dims(query, 1)
        score = self.V(tf.nn.tanh(self.W1(query) + self.W2(values)))
        attention_weights = tf.nn.softmax(score, axis=1)
        context_vector = attention_weights * values
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector, attention_weights

def build_attention_model():
    encoder_inputs = Input(shape=(MAX_INPUT_LEN,))
    enc_emb = Embedding(VOCAB_SIZE, EMBEDDING_DIM)(encoder_inputs)
    enc_out, state_h, state_c = LSTM(LSTM_UNITS, return_sequences=True, return_state=True)(enc_emb)

    decoder_inputs = Input(shape=(MAX_TARGET_LEN-1,))
    dec_emb = Embedding(VOCAB_SIZE, EMBEDDING_DIM)(decoder_inputs)
    dec_lstm_out = LSTM(LSTM_UNITS, return_sequences=True)(dec_emb, initial_state=[state_h, state_c])

    attention = BahdanauAttention(LSTM_UNITS)
    context_vector, _ = attention(state_h, enc_out)
    repeated_context = RepeatVector(MAX_TARGET_LEN-1)(context_vector)
    concat = Concatenate(axis=-1)([dec_lstm_out, repeated_context])
    output = Dense(VOCAB_SIZE, activation='softmax')(concat)

    model = Model([encoder_inputs, decoder_inputs], output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# STEP 6: Model Training

def train_model(model, model_name, X_train, Xd_train, Y_train, X_val, Xd_val, Y_val):
    print(f"\nTraining {model_name}...")

    callbacks = [
        tf.keras.callbacks.EarlyStopping(patience=3, monitor='val_loss', restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(factor=0.1, patience=2)
    ]

    history = model.fit(
        [X_train, Xd_train], Y_train,
        validation_data=([X_val, Xd_val], Y_val),
        batch_size=64,
        epochs=20,
        callbacks=callbacks,
        verbose=1
    )

    plt.figure(figsize=(10, 4))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Training Progress')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

    return model

vanilla_model = train_model(build_lstm_model(), 'Vanilla LSTM', X_train, Xd_train, Y_train, X_val, Xd_val, Y_val)
attention_model = train_model(build_attention_model(), 'LSTM + Attention', X_train, Xd_train, Y_train, X_val, Xd_val, Y_val)

# STEP 7: Inference

def generate_headline(input_text, model, input_tokenizer, target_tokenizer, temperature=1.0):
    input_seq = input_tokenizer.texts_to_sequences([clean_text(input_text)])
    input_seq = pad_sequences(input_seq, maxlen=MAX_INPUT_LEN, padding='post')

    start_token = target_tokenizer.word_index['start']
    decoder_input = np.zeros((1, MAX_TARGET_LEN-1))
    decoder_input[0, 0] = start_token

    output = []
    for i in range(MAX_TARGET_LEN-1):
        predictions = model.predict([input_seq, decoder_input], verbose=0)[0, i]
        predictions = np.log(predictions + 1e-10) / temperature
        exp_preds = np.exp(predictions)
        predictions = exp_preds / np.sum(exp_preds)
        sampled_token_index = np.random.choice(len(predictions), p=predictions)
        sampled_word = target_tokenizer.index_word.get(sampled_token_index, '')
        if sampled_word == 'end' or sampled_word == '<OOV>':
            break
        output.append(sampled_word)
        if i+1 < MAX_TARGET_LEN-1:
            decoder_input[0, i+1] = sampled_token_index
    return ' '.join(output)

# STEP 8: Evaluation

def evaluate_models(models, model_names, X_val, Y_val, input_tokenizer, target_tokenizer):
    results = pd.DataFrame(columns=['BLEU-1', 'BLEU-4', 'ROUGE-L', 'METEOR', 'Accuracy'])

    references = []
    for seq in Y_val:
        ref_words = []
        for token in seq:
            if token[0] > 0:
                word = target_tokenizer.index_word.get(token[0], '')
                if word and word != '<OOV>' and word != 'end':
                    ref_words.append(word)
        references.append(' '.join(ref_words))

    for name, model in zip(model_names, models):
        predictions = []
        for i in range(len(X_val)):
            input_text = ' '.join([input_tokenizer.index_word.get(token, '')
                                  for token in X_val[i] if token > 0 and input_tokenizer.index_word.get(token, '') != '<OOV>'])
            predictions.append(generate_headline(input_text, model, input_tokenizer, target_tokenizer))

        ref_tokens = [ref.split() for ref in references]
        pred_tokens = [pred.split() for pred in predictions]

        bleu1 = np.mean([sentence_bleu([ref], pred, weights=(1, 0, 0, 0)) for ref, pred in zip(ref_tokens, pred_tokens)])
        bleu4 = np.mean([sentence_bleu([ref], pred, weights=(0.25, 0.25, 0.25, 0.25)) for ref, pred in zip(ref_tokens, pred_tokens)])

        scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
        rouge_l = np.mean([scorer.score(ref, pred)['rougeL'].fmeasure for ref, pred in zip(references, predictions)])

        meteor = np.mean([meteor_score([ref.split()], pred.split()) for ref, pred in zip(references, predictions)])

        accuracy = np.mean([1 if pred == ref else 0 for pred, ref in zip(predictions, references)])

        results.loc[name] = [bleu1, bleu4, rouge_l, meteor, accuracy]
    return results

metrics_results = evaluate_models(
    [vanilla_model, attention_model],
    ['Vanilla LSTM', 'LSTM + Attention'],
    X_val, Y_val, input_tokenizer, target_tokenizer
)

print("\nDetailed Metrics Results:")
print(metrics_results)

metrics_results.plot(kind='bar', figsize=(12, 6))
plt.title('Model Comparison Metrics')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.show()
```
